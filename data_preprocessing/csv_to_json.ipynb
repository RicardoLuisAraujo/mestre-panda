# Answers

```python
answers_df = pd.melt(df, id_vars=['ID', 'true_answer'], value_vars=['option_1', 'option_2', 'option_3', 'option_4'], var_name='answer_option', value_name='answer')
answers_df.rename(columns={'ID': 'exercise_id', 'true_answer': 'real_answer'}, inplace=True)

answers_df.head()
```

```python
answers_df.sort_values(by=['exercise_id'], inplace=True)
answers_df.reset_index(drop=True, inplace=True)
# answers_df['real_answer'] = answers_df['real_answer'].str.replace(',', ', ')
answers_df.head(10)
```

```python
answers_df['true_answer'] = np.where((answers_df['real_answer'] == answers_df['answer']), True, False)
```

```python
answers_df.head()
```

```python
answers_df.dropna().head()
answers_df['exercise_number'] = answers_df['exercise_id']
```

```python
answers_df[['exercise_number', 'true_answer', 'answer']].dropna().to_json('answers.json', orient='records')
```
